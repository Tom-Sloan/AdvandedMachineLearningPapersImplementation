```python
import os
import shutil
from tqdm import tqdm
from pathlib import Path

def reorganize_segmentation_dir():
    # Base paths
    base_dir = 'Stoneflies'
    src_dir = os.path.join(base_dir, 'segmentation')
    
    # Create temporary directory for reorganization
    temp_dir = os.path.join(base_dir, 'segmentation_temp')
    os.makedirs(temp_dir, exist_ok=True)
    
    # Process each set (set0, set1)
    for set_name in ['set0', 'set1']:
        set_path = os.path.join(src_dir, set_name)
        if not os.path.isdir(set_path):
            continue
            
        # Process each bug type directory
        for bug_type in tqdm(os.listdir(set_path), desc=f"Processing {set_name}"):
            bug_type_path = os.path.join(set_path, bug_type)
            if not os.path.isdir(bug_type_path):
                continue
                
            # Create new directory structure
            new_bug_dir = os.path.join(temp_dir, bug_type, set_name)
            os.makedirs(new_bug_dir, exist_ok=True)
            
            # Move all files
            for file_name in os.listdir(bug_type_path):
                src_file = os.path.join(bug_type_path, file_name)
                dst_file = os.path.join(new_bug_dir, file_name)
                shutil.copy2(src_file, dst_file)
    
    # Remove old segmentation directory and rename temp to segmentation
    shutil.rmtree(src_dir)
    os.rename(temp_dir, src_dir)
    print("Segmentation directory reorganized successfully!")

# Run the reorganization
reorganize_segmentation_dir()
```

```python
def reorganize_stonefly_dir():
    # Base paths
    base_dir = 'Stoneflies'
    src_dir = os.path.join(base_dir, 'stonefly')
    
    # Create temporary directory for reorganization
    temp_dir = os.path.join(base_dir, 'stonefly_temp')
    os.makedirs(temp_dir, exist_ok=True)
    
    # Process each bug type directory
    for bug_type in tqdm(os.listdir(src_dir), desc="Processing stonefly"):
        bug_type_path = os.path.join(src_dir, bug_type)
        if not os.path.isdir(bug_type_path):
            continue
            
        # Process each set directory
        for set_name in os.listdir(bug_type_path):
            set_path = os.path.join(bug_type_path, set_name)
            if not os.path.isdir(set_path):
                continue
                
            # Create new directory structure
            new_set_dir = os.path.join(temp_dir, set_name, bug_type)
            os.makedirs(new_set_dir, exist_ok=True)
            
            # Move all files
            for file_name in os.listdir(set_path):
                src_file = os.path.join(set_path, file_name)
                dst_file = os.path.join(new_set_dir, file_name)
                shutil.copy2(src_file, dst_file)
    
    # Remove old stonefly directory and rename temp to stonefly
    shutil.rmtree(src_dir)
    os.rename(temp_dir, src_dir)
    print("Stonefly directory reorganized successfully!")

# Run the reorganization
reorganize_stonefly_dir()
```
